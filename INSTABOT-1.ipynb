# Question 1 : Login to your Instagram
Login to your Instagram Handle
Submit with sample username and password

```python
def login(username, password):
    try:
        #Entering Username
        enter_username = driver.find_element_by_name('username')
        enter_username.send_keys(username)
        #Entering Password
        enter_password = driver.find_element_by_name('password')
        enter_password.send_keys(password)
        #Logging In
        login = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="loginForm"]/div[1]/div[3]/button/div')))
        login.click()
        #Dealing with save login info page
        save_login_info = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/div/div/section/div/button')))
        save_login_info.click()
        #Dealing with turn on notifications page
        turn_on_notifications = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//button[@class="aOOlW   HoLwm "]')))
        turn_on_notifications.click()
    except TimeoutException :
        print("Something went wrong! Try Again")
        
```

## Q2) Type for ‚Äúfood‚Äù in search bar and print all the names of the Instagram Handles that are displayed in list after typing ‚Äúfood‚Äù

```python
def search(name):
    #Getting search bar
    search = driver.find_element_by_class_name('XTCLo')
    search.send_keys(name)
    handles = []
    time.sleep(3)
    names = driver.find_elements_by_xpath('//div[@class="fuqBx "]/div/a["href"]')
    for i in names:
        if i.text[0]!='#':
            handles.append(i.text.split('\n')[0])
    return handles
```

```python
handles = search('food')
for i in handles:
    print(i)
```

# Question 4 : Follow/Unfollow given handle
Follow/Unfollow given handle -
1.Open the Instagram Handle of ‚ÄúSo Delhi‚Äù
2.Start following it. Print a message if you are already following
3.After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def follow():
    follow_button = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//button[@class="_5f5mN       jIbKX  _6VtSN     yZn4P   "]')))
    if(follow_button.text=="Follow"):
        follow_button.click()
    else:
        print("Already followed!")

def unfollow():
    unfollow_button = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'//button[@class="_5f5mN    -fzfL     _6VtSN     yZn4P   "]')))
    
    if(unfollow_button.text != "Follow"):
        unfollow_button.click()
        unfollow_button_conformation = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'//button[@class="aOOlW -Cab_   "]')))
        unfollow_button_conformation.click()
    else:
        print("Already unfollowed!")

follow()
unfollow()
```

# Question 5 : Like/Unlike posts
Like/Unlike posts
1.Liking the top 30 posts of the ‚Äòdilsefoodie'. Print message if you have already liked it.
2.Unliking the top 30 posts of the ‚Äòdilsefoodie‚Äô. Print message if you have already unliked it

```python
def like():
    driver.execute_script('window.scrollTo(0, 6000);')
    time.sleep(3)
    driver.execute_script('window.scrollTo(0, -6000);')
    time.sleep(3)
    try:
        posts = driver.find_elements_by_xpath('//div[@class="v1Nh3 kIKUG  _bz0w"]/a')
        for i in range(30):
            posts[i].click()
            time.sleep(4)
            like = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME,'fr66n')))
            liked = bs(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            if liked=='Like':
                like.click()
                time.sleep(2)
            else:
                print("You have already liked the post number", i+1)
                time.sleep(2)
            back = driver.find_elements_by_class_name('wpO6b  ')
            back[2].click()
            time.sleep(1)
    except TimeoutException:
        print('Something went wrong')

def dislike():
    try:
        posts = driver.find_elements_by_xpath('//div[@class="v1Nh3 kIKUG  _bz0w"]/a')
        for i in range(30):
            posts[i].click()
            time.sleep(4)
            unlike = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME,'fr66n')))
            unliked = bs(unlike.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            if unliked == "Unlike":
                unlike.click()
                time.sleep(2)
            else:
                print("You have already unliked the post number", i+1)
                time.sleep(2)
            back = driver.find_elements_by_class_name('wpO6b  ')
            back[2].click()
            time.sleep(1)
    except TimeoutException:
        print('Something went wrong')
```

# Problem 7 : Check the story of ‚Äòcoding.ninjas‚Äô
Check the story of ‚Äòcoding.ninjas‚Äô. Consider the following Scenarios and print error messages accordingly -
1.If You have already seen the story.
2.Or The user has no story.
3.Or View the story if not yet seen.

```python
open_profile('coding.ninjas')
```

```python
def Check_Story():
    try:
        #locating story or profile pic
        story = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME,"RR-M-.h5uC0")))
    
        #check the Profile photo size to find out story is available or not
        height = driver.find_element_by_class_name('CfWVH').get_attribute('height')
        if int(height) == 208:
            print("Already seen the story")
        else:
            print("Viewing the story")
            driver.execute_script("arguments[0].click();",story)
    except:
        print("No story is available to view")
```
